```python
import subprocess

cpp_executable = "./practice"

test_cases = [
    (0, 0, 0, "Infinite solutions"),
    (0, 0, 5, "No solution"),
    (0, 2, -8, "2 solutions"),
    (1, 0, -1, "2 solutions"),
    (1, -5, 4, "4 solutions"),
    (1, 2, 5, "No solution"),
]

for i, (a, b, c, expected) in enumerate(test_cases, 1):
    proc = subprocess.Popen(
        [cpp_executable],
        stdin=subprocess.PIPE,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )
    input_str = f"{a}\n{b}\n{c}\n"
    out, err = proc.communicate(input=input_str)

    out_clean = out.strip()

    result = "PASS" if expected in out_clean else "FAIL"
    print(f"Test {i}: {result}")
    print(f"  Input: a={a}, b={b}, c={c}")
    print(f"  Expected : '{expected}'")
    print(f"  Actual : '{out_clean}'")
    if err:
        print(f"  STDERR: '{err.strip()}'")
    print("-" * 50)
```
